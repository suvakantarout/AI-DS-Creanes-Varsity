```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
data = pd.read_csv('housing.csv')
df = pd.DataFrame(data)
df.head()
```

```python
df.describe() #Summary about dataframe
```

```python
df.info() # structure of dataset
```

```python
df.isnull().sum()  # Identify how many missing valus are present in each column of the data frame
```

```python
df_encoded = pd.get_dummies(df, drop_first=True)
x = df_encoded[['median_income']]  
y = df_encoded['median_house_value']  
print(x.head())
print(y.head())
```

```python
# Splits the data into 80% training and 20% testing.
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
model = LinearRegression()  #Creating a linear regression model.
model.fit(X_train, y_train) #Training the mode
y_pred = model.predict(X_test) #Predict the values
```

```python
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([y.min(), y.max()], [y.min(), y.max()],'--r')
plt.plot()
plt.show()
```
