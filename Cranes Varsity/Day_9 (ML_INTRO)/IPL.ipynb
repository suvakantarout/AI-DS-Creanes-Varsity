# Generate Bowler Statistics from IPL Dataset
You must calculate the following metrics for each bowler:

- Bowler
- Innings – Number of unique matches bowled
- Balls Bowled – Count of valid deliveries
- Runs Conceded – Total runs given (including boundaries and extras)
- Wickets – Dismissals excluding run-outs and retired hurt
- Dot Balls – Deliveries where 0 runs were scored (excluding extras)
- 4s Conceded – Number of times 4 runs were conceded
- 6s Conceded – Number of times 6 runs were conceded
- Overs – Balls Bowled / 6 (rounded to 1 decimal)
- Economy – Runs Conceded / Overs
- Strike Rate – Balls Bowled / Wickets
- Average – Runs Conceded / Wickets
